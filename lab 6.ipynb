```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/Users/jackmoskowitz/PycharmProjects/Lab6/.venv/company_sales_data.csv')

month_numbers = data['month_number']
total_profits = data['total_profit']
plt.figure(figsize=(10, 6))
plt.plot(month_numbers, total_profits, marker='o', color='b', linestyle='-')
plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(range(1, 13))

plt.show()
```

```python
data = pd.read_csv('/Users/jackmoskowitz/PycharmProjects/Lab6/.venv/company_sales_data.csv')


face_cream_sales = data['facecream']
face_wash_sales = data['facewash']
month_numbers = data['month_number'] 


plt.figure(figsize=(10, 6))
plt.bar(month_numbers - 0.2, face_cream_sales, width=0.4, color='blue', label='Face Cream')
plt.bar(month_numbers + 0.2, face_wash_sales, width=0.4, color='orange', label='Face Wash')

plt.title('Face Cream and Face Wash Sales per Month', fontsize=16)
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Units Sold', fontsize=14)
plt.xticks(month_numbers)
plt.legend(fontsize=12)
plt.grid(True)

plt.show()
```

```python
data = pd.read_csv('/Users/jackmoskowitz/PycharmProjects/Lab6/.venv/company_sales_data.csv')

bathing_soap_sales = data['bathingsoap']
facewash_sales = data['facewash']
month_numbers = data['month_number']


fig, axes = plt.subplots(2, 1, figsize=(10, 10))


axes[0].plot(month_numbers, bathing_soap_sales, marker='o', color='black')
axes[0].set_title('Bathing Soap Sales per Month')
axes[0].set_xlabel('Month Number')
axes[0].set_ylabel('Units Sold')


axes[1].plot(month_numbers, facewash_sales, marker='o', color='red')
axes[1].set_title('Facewash Sales per Month')
axes[1].set_xlabel('Month Number')
axes[1].set_ylabel('Units Sold')


plt.tight_layout()


plt.show()
```

```python
data = pd.read_csv('/Users/jackmoskowitz/PycharmProjects/Lab6/.venv/company_sales_data.csv')

products = data.columns[1:-2]
sales_data = data[products]

plt.figure(figsize=(10, 6))
plt.stackplot(data['month_number'], sales_data.T, labels=products)

plt.title('Product Sales Data Over Time', fontsize=16)
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Units Sold', fontsize=14)
plt.legend(loc='upper left', fontsize=12)


plt.show()
```
